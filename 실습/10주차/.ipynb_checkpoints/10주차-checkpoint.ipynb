```python
date_list = []
low_list = []
high_list = []
rain_list = []
```

```python
for data in data_list:
    temp = data.split(',')
    date_list.append(temp[1])
    low_list.append(float(temp[2]))
    high_list.append(float(temp[3]))
    if temp[4] == '\n':
        rain_list.append(0)
    else:
        rain_list.append(float(temp[4]))
```

```python
class FourCal:
    def __init__(self, first = 1, second = 1):
        self.first = first
        self.second = second
    
    def setdata(self, first, second):
        self.first = first
        self.second = second
    
    def sum(self):
        result = self.first + self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal(4, 2)
b = FourCal(5, 1)
c = FourCal()
```

```python
a.setdata(4, 2)
b.setdata(5, 1)
```

```python
print(a.first, b.first)
```

```python
print(a.sum())
```

```python
print(a.div())
```

```python
print(a.sub())
```

```python
print(c.sub())
```

```python
print(c.sum())
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
```

```python
a = MoreFourCal(4, 2)
a.sum()
```

```python
a.mul()
```

```python
a.pow()
```

```python
class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
```

```python
a = SafeFourCal(4, 0)
```

```python
a.div()
```
